```python
# H5 py file T to list 
h5_files = {}
train_list_1 = []
train_list_2 = []
train_list_3 = []
train_list_4 = []
id_list_1 = []
def get_h5(h5_path):
        # checks if h5 path is already in dictionary of h5py file paths
        if h5_path in h5_files:
            # returns them if there already there
            return h5_files[h5_path]
        else:
            # Sets the path as a key, and the value as the read file
            h5_files[h5_path] = h5py.File(h5_path, "r")

index = range(0,len(frag_tn5_train_df))
print(index)
for i in index:
    bam_file_index = (index[i])
    get_h5(frag_tn5_train_df.iloc[bam_file_index,frag_tn5_train_df.columns.get_loc('diff_frag_h5_path')])
    id_val_1 = frag_tn5_train_df.iloc[ bam_file_index, frag_tn5_train_df.columns.get_loc('patient_id')]
    h5_tn5_read_count_diff_1 = h5_files.get(frag_tn5_train_df.iloc[ bam_file_index, frag_tn5_train_df.columns.get_loc('diff_frag_h5_path')])
    T_1 = h5_tn5_read_count_diff_1.attrs["total_sum"]
    
    get_h5(frag_tn5_train_df.iloc[bam_file_index,frag_tn5_train_df.columns.get_loc('int_gamma_diff_frag_h5_path')])
    h5_tn5_read_count_diff_2 = h5_files.get(frag_tn5_train_df.iloc[ bam_file_index, frag_tn5_train_df.columns.get_loc('int_gamma_diff_frag_h5_path')])
    T_2 = h5_tn5_read_count_diff_2.attrs["total_sum"]
    
    get_h5(frag_tn5_train_df.iloc[bam_file_index,frag_tn5_train_df.columns.get_loc('salm_int_gamma_diff_frag_h5_path')])
    h5_tn5_read_count_diff_3 = h5_files.get(frag_tn5_train_df.iloc[ bam_file_index, frag_tn5_train_df.columns.get_loc('salm_int_gamma_diff_frag_h5_path')])
    T_3 = h5_tn5_read_count_diff_3.attrs["total_sum"]
    
    get_h5(frag_tn5_train_df.iloc[bam_file_index,frag_tn5_train_df.columns.get_loc('salm_diff_frag_h5_path')])
    h5_tn5_read_count_diff_4 = h5_files.get(frag_tn5_train_df.iloc[ bam_file_index, frag_tn5_train_df.columns.get_loc('salm_diff_frag_h5_path')])
    T_4 = h5_tn5_read_count_diff_4.attrs["total_sum"]
    
    #print('T_1',T_1)
    id_list_1.append(id_val_1)
    train_list_1.append(T_1)
    train_list_2.append(T_2)
    train_list_3.append(T_3)
    train_list_4.append(T_4)
```

```python
train_df = pd.DataFrame()
train_df['ID'] = id_list_1
train_df['T_1_diff'] = train_list_1
train_df['T_2_int_gamma'] = train_list_2
train_df['T_3_salm_int_gamma'] = train_list_3
train_df['T_4_salm_diff'] = train_list_4
```
